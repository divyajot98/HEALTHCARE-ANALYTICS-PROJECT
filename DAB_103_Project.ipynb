```python
# Check for null values in the DataFrame
null_values = train.isnull().sum()

# Print the number of null values in each column
print(null_values)

#we use the isnull() function to check for null values in the DataFrame. Then, we print the number of null values in each column using the sum() function on the resulting DataFrame. This will print a series object with the sum of null values for each column in the DataFrame.
```

```python
# Check for incomplete records in the DataFrame
incomplete_records = train[train.isnull().any(axis=1)]

# Print the incomplete records
print(incomplete_records)

#we use the isnull() function to check for null values in the DataFrame, and the any() function with axis=1 to check for rows that contain at least one null value (i.e., incomplete records). Finally, we print the incomplete records using boolean indexing on the original DataFrame. This will print all the rows that contain at least one null value.
```

```python
#Replacing NA values in Bed Grade Column for both Train datasets
train['Bed Grade'].fillna(train['Bed Grade'].mode()[0], inplace = True)
#test['Bed Grade'].fillna(test['Bed Grade'].mode()[0], inplace = True)
```

```python
#Replacing NA values in  Column for both Train datasets
train['City_Code_Patient'].fillna(train['City_Code_Patient'].mode()[0], inplace = True)
```

```python
# Label Encoding Stay column in train dataset
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['Stay'] = le.fit_transform(train['Stay'].astype('str'))
```
