```python
#Replacing NA values in Bed Grade Column for both Train datasets
train['Bed Grade'].fillna(train['Bed Grade'].mode()[0], inplace = True)
```

```python
#Replacing NA values in  Column for both Train datasets
train['City_Code_Patient'].fillna(train['City_Code_Patient'].mode()[0], inplace = True)
```

```python
train.head()
```

```python
#import pandas as pd
train.rename(columns={
    'Bed Grade': 'Bed_Grade', 
    'patientid ': 'patient_id ', 
    'Type of Admission' : 'Type_of_Admission',
    'Severity of Illness' : 'Severity_of_Illness',
    'Visitors with Patient' : 'Visitors_with_Patient',
    'Available Extra Rooms in Hospital' : 'Available_Rooms',
}, inplace=True)
train 
```

```python
#Dropping the rows with incorrect values from the Age column i.e = '20-Nov'
train = train.drop(train[train.Age == '20-Nov'].index)
train
```

```python
#Replace the incorrect timestamp values with integer values in Stay column 
train.Stay.replace('20-Nov','0-10',inplace=True)
```

```python
#Replace the incorrect timestamp values with integer values in Stay column 
train.Stay.replace('More than 100 days','100+',inplace=True)
```

```python
import matplotlib.pyplot as plt

# Filter the data to remove the surgery department
filtered_data = train[train['Department'] != 'surgery']

# Count the number of occurrences of each department
dept_counts = filtered_data['Department'].value_counts()

# Set the explode value for the department with more counts
explode = [0.1 if count == dept_counts.max() else 0 for count in dept_counts.values]

# Create a pie chart
plt.pie(dept_counts.values, labels=dept_counts.index, autopct='%1.1f%%', explode=explode)

# Add a title
plt.title('Department Distribution')

# Show the plot
plt.show()
```
